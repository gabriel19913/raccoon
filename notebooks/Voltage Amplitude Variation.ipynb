# Voltage Amplitude Variation (Figure)

```python
import sys
sys.path.append('..')
```

## Load Data

```python
import wfdb

signal_path = "../data/mitdb/105"
record = wfdb.rdrecord(signal_path)
```

## Plot

```python
import matplotlib.pyplot as plt
plt.rcParams.update({"font.size": 16})

fig, ax = plt.subplots()
fig.set_size_inches(15,5)

ax.plot(record.p_signal[:,0])
ax.set_xlabel("Time / Samples")
ax.set_ylabel("Voltage / mV")
ax.autoscale(enable=True, axis='x', tight=True)
```

## Save Plot

```python
save = False

if save:
    fig.savefig(
        '../../msc-thesis/thesis/images/voltage_amplitude_variation.pdf',
        format='pdf',
        bbox_inches = 'tight',
        pad_inches = 0.1)
```
